```python
import matplotlib.pyplot as plt
import rng
```

### Middle Square Method

```python
rng.set_generator('middle-square', 57)

sample = rng.uniform(0, 1, size=10)

_ = plt.hist(sample)
```

Middle Square is not very robust pseudo random generator, it falls into a short cycle very fast (usually in less than ~50 iterations)

### Linear congruence generator

```python
rng.set_generator('congruence', 57)

sample = rng.uniform(size=10_000)
_ = plt.hist(sample, density=True, bins=50)

x = []
y = []
for i in range(len(sample) // 2):
    x.append(sample[i])
    y.append(sample[ i + 1])

plt.figure()
plt.scatter(x, y)
```

### Mersenne Twister (MT19937)

```python
rng.set_generator('mt19937', 57)

sample = rng.uniform(size=10_000)
_ = plt.hist(sample, density=True, bins=50)

x = []
y = []
for i in range(len(sample) // 2):
    x.append(sample[i])
    y.append(sample[ i + 1])

plt.figure()
plt.scatter(x, y)
```
