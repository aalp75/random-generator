```python
import rng
import math
```

### $\chi^2$ Test of Homogeneity 

```python
def poisson_density(x, lamb=1):
    return (math.exp(-lamb) * pow(lamb,x) / math.factorial(x))

sample = rng.poisson(1, size=10_000)

T = rng.chi2_test(sample, lambda x : poisson_density(x, 1), alpha=0.05)
print("T =",T)

T = rng.chi2_test(sample, lambda x : poisson_density(x, 2), alpha=0.05)
print("T =",T)
```

## $\chi^2$ Test of Independance

```python
x = rng.uniform_discrete(1, 6, size=10_000)
y = rng.uniform_discrete(1, 6, size=10_000)

T = rng.indep_test(x, x)
print("T =",T)

T = rng.indep_test(x, y)
print("T =",T)
```
