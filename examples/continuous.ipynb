```python
import matplotlib.pyplot as plt
import rng
```

### Uniform [a, b]

```python
sample = rng.uniform(5, 12, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

### Exponential

```python
sample = rng.exponential(1.5, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

### Gaussian

##### Scalar

```python
sample = rng.gaussian(2, 2, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

#### Vector

```python
sample = rng.gaussian_vector([[2, 3]], [[1, 0.5], [0.5, 1]], n=2)
```

### Pareto

```python
sample = rng.pareto(1, 6, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

### Chi-square

```python
sample = rng.chisquare(2, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```
