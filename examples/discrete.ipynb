```python
import matplotlib.pyplot as plt
import rng
```

### Bernouilli

```python
sample = rng.bernouilli(0.3, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

### Uniform discrete

```python
sample = rng.uniform_discrete(1, 6, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

### Binomial

```python
sample = rng.binomial(10, 0.3, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```

```python
sample = rng.geometric(0.2, size=10_000)
_ = plt.hist(sample, density=True, bins=50)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```
