```python
import matplotlib.pyplot as plt
import rng
```

### Bernouilli

```python
sample = rng.bernouilli(0.3, size=1_000)
_ = plt.hist(sample, density=True, bins=3)
expectation = rng.expectation(sample)
variance = rng.variance(sample)
print(f'Expectation: {expectation}')
print(f'Variance: {variance}')
```
